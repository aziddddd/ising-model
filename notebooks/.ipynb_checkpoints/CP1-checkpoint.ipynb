#### Import required packages

```python
%matplotlib notebook

import im_utils
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML, clear_output

plt.ioff()
```

#### Initialisation

```python
N           = 50                             # System size
temp        = .4
method      = 'glauber'                      # glauber or kawasaki
```

#### Methods

```python
ising_model = im_utils.Ising(N, temp, method)
```

#### Results

```python
plotFigure = plt.figure(figsize=(5, 5), dpi=80)
anim = animation.FuncAnimation(plotFigure, ising_model.simulate, frames=300, interval=10, repeat=False)
```

```python
HTML(anim.to_html5_video())
```

```python
anim.save("test.mp4", writer='ffmpeg')
```
