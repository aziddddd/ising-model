
#### Import required packages

```python
import im_utils
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
plt.rcParams['figure.figsize'] = (10,10)
```

#### Initialisation

```python
N           = 50                              # System size
temp        = .4
method      = 'kawasaki'                      # glauber or kawasaki
temp_point  = 88
temp_range  = (1.53, 3.28)
equistep    = 100
calcstep    = 10
```

#### Methods

```python
ising_model = im_utils.Ising(N, temp, method, temp_point, temp_range, equistep, calcstep)
```

#### Results

```python
ising_model.snapshots(1000)
```

- Dynamic plotting ( Smoother)

```python
ising_model.DynamicPlot(100)
```

- Animation

```python
%matplotlib notebook
plotFigure = plt.figure(figsize=(5, 5), dpi=80)
animate = animation.FuncAnimation(plotFigure, ising_model.simulate, interval=1, repeat=False)
```
